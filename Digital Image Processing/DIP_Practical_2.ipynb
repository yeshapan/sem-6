**Objective:**

Write a program to perform following image processing operations:
1.   Reading an image and converting into grayscale image
2.   Resize the given image
3.   Plot the histogram of a given image
4.   Write a text on an image
5.   Resize the given image with aspect ratio
6.   Display various image properties


```python
#import necessary libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

**1. Read Image and convert to grayscale**

```python
#read image

#define the image path
image_path = 'hogwarts image.jpg'

#read the original image
#cv2.imread loads the image in BGR format by default
original_img = cv2.imread(image_path)

#display the original image
cv2_imshow(original_img)
```

```python
#covert to grayscale
#cv2.COLOR_BGR2GRAY removes color channels (leaving only intensity data)
gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

#display gray image
cv2_imshow(gray_img)
```

**2. Resize Image (Fixed Dimension)**

```python
#resize image to fixed dimensions

#resizing to exactly (300x300 pixels) ignores original aspect ratio (may squash/stretch image)
resized_fixed = cv2.resize(original_img, (300, 300))

#display resized image
cv2_imshow(resized_fixed)
```

**3. Plot Histogram of Image**

```python
#plot histogram
'''
histogram shows the distribution of pixel intensities (0 to 255)
Arguments: [image], [channel], mask, [histSize], [ranges]
'''
hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

#use matplotlib to plot graph
plt.figure(figsize=(10, 5))
plt.title("Grayscale Histogram (Pixel Intensity Distribution)")
plt.xlabel("Pixel Value (0=Black, 255=White)")
plt.ylabel("Frequency (Pixel Count)")
plt.plot(hist, color='black')
plt.xlim([0, 256])
plt.grid(True, alpha=0.3)
plt.show()
```

**4. Write text on Image**

```python
#write text on image
#copy original image
img_with_text = original_img.copy()

#text settings
text = "Hogwarts School of Witchcraft and Wizardry"
position = (40, 50) # x, y coordinates
font = cv2.FONT_HERSHEY_COMPLEX
font_scale = 0.6
color = (0, 0, 0)
thickness = 1

#apply text
cv2.putText(img_with_text, text, position, font, font_scale, color, thickness)

cv2_imshow(img_with_text)
```

**5. Resize given image with aspect ratio**

```python
#we want to scale the image down to 50% of its original size
scale_percent = 50

#calculate new dimensions based on the percentage
new_width = int(original_img.shape[1] * scale_percent / 100)
new_height = int(original_img.shape[0] * scale_percent / 100)
new_dim = (new_width, new_height)

#resize using the calculated dimensions
'''
INTER_AREA is the preferred interpolation method for shrinking images
'''
resized_aspect = cv2.resize(original_img, new_dim, interpolation=cv2.INTER_AREA)

print(f"Original Dimensions: {original_img.shape[1]}x{original_img.shape[0]}")
print(f"New Dimensions (50%): {new_width}x{new_height}")

#display scaled image
cv2_imshow(resized_aspect)
```

**6. Display various image properties**

```python
#shape: Returns a tuple (Rows/Height, Columns/Width, Channels)
print(f"1. Image Shape: {original_img.shape}")
print(f"   - Height: {original_img.shape[0]} pixels")
print(f"   - Width:  {original_img.shape[1]} pixels")
print(f"   - Channels: {original_img.shape[2]} (3 for RGB/BGR)")

#size: Total number of pixels (Height * Width * Channels)
print(f"2. Total Pixel Count: {original_img.size}")

#dtype: Data type of pixel values (usually uint8, meaning 0-255)
print(f"3. Data Type: {original_img.dtype}")
```
