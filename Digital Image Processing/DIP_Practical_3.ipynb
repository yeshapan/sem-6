**Objective:**

Write and execute a program for image morphological operations - erosion and dilation.

**How they work:**

Morphological operations are non-linear transformations that modify the geometric structure of an image. They rely on a Structuring Element (SE), which is a small matrix (kernel) that slides over the image.

1. **Erosion ($A \ominus B$)**

* Shrinks the foreground (white pixels)
* A pixel remains 1 (white) only if all pixels under the kernel are 1. If even one pixel is 0 (black), the center pixel is set to 0
* Removes small noise (salt noise), detaches connected objects, and thins structures

2. **Dilation ($A \oplus B$)**
* Expands the foreground (white pixels)
* A pixel becomes 1 (white) if at least one pixel under the kernel is 1
* Fills small holes (pepper noise), joins broken segments, and thickens structures.

**1. Import libraries and image**

```python
#import libraries needed
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#load and show original image
img=cv2.imread('hogwarts image.jpg')
cv2_imshow(img)
```

**Pre-req: convert image to grayscale**

```python
#convert to grayscale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

**2. Binary Thresholding and Structural Element**

```python
#convert to binary
'''
convert to pure Black (0) and White (255) to separate image from background
'''
_, binary = cv2.threshold(gray, 115, 255, cv2.THRESH_BINARY)


cv2_imshow(binary)
```

```python
#create a 4x4 Matrix of 1s. This is our 'probe' or 'kernel'.
kernel_size = 4
kernel = np.ones((kernel_size, kernel_size), np.uint8)
print(f"Structuring Element ({kernel_size}x{kernel_size} Kernel)")
print(kernel)
```

**3. Erosion Operation**

```python
'''
Rule: A pixel is 1 only if ALL pixels under the kernel are 1
Result: Objects shrink, small white noise is removed
'''
erosion = cv2.erode(binary, kernel, iterations=1)

print("Erosion Result (Shrinking)")
cv2_imshow(erosion)
```

**4. Dilation Operation**

```python
'''
Rule: A pixel is 1 if AT LEAST ONE pixel under the kernel is 1
Result: Objects grow, small black holes are filled.
'''
dilation = cv2.dilate(binary, kernel, iterations=1)

print("Dilation Result (Expanding)")
cv2_imshow(dilation)
```

The practical demonstrated that:
* Erosion is effective for removing small white noise (salt noise) and detaching connected objects
* Dilation is effective for filling small holes (pepper noise) and bridging gaps between objects.

The size of the kernel determines the magnitude of the operation.
